```python
import pandas as pd
```

```python
train_df   = pd.read_csv('dbpedia_csv/train.csv', names=('class', 'title', 'abstract'))
valid_df   = pd.read_csv('dbpedia_csv/test.csv' , names=('class', 'title', 'abstract'))
```

```python
class_sr = pd.read_csv('dbpedia_csv/classes.txt', header=None)
class_sr.index = class_sr.index + 1
class_dict = class_sr.to_dict()[0]
class_sr
```

```python
train_df['class'] = train_df['class'].map(class_dict)
valid_df['class'] = valid_df['class'].map(class_dict)
train_df
```
