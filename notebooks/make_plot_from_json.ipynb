```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dir = '../reports/csv'
for val in ('loss', 'accu'):

    word2vec = pd.read_csv(f'{dir}/run-word2vec_bs128_valid-tag-{val}.csv')['Value'].tolist()
    bert_emb = pd.read_csv(f'{dir}/run-bert-emb_bs128_valid-tag-{val}.csv')['Value'].tolist()
    bert     = pd.read_csv(f'{dir}/run-bert-base_bs32_valid-tag-{val}.csv')['Value'].tolist()

    if val == 'accu': val = 'accuracy'
    plt.plot(range(1, 31), word2vec, label='word2vec + RNN' )
    plt.plot(range(1, 31), bert_emb, label='BERTembed + RNN' )
    plt.plot(range(1, 31), bert    , label='BERT-base')
    plt.xlabel('epoch')
    plt.ylabel(val)
    plt.legend()
    plt.show()
```
